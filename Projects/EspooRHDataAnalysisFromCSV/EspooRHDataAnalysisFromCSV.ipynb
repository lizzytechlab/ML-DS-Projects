```python
# Loading data from CSV
weather_data = pd.read_csv('weather_2021-12.csv')
```

```python
# Filtering and extracting average relative humidity values
rh_values_extracted = weather_data[weather_data['ParameterName'] == 'RH_PT1H_AVG']['ParameterValue']
```

```python
# Viewing a snapshot from the extracted data.
rh_values_extracted.head(10)
```

```python
# Dropping NaN values from the extracted relative humidity data
valid_rh_values = rh_values_extracted.dropna()

# Calculating the average relative humidity
average_rh = valid_rh_values.mean()
```

```python
# Displaying calculated average relative humidity
print("Average Relative Humidity Calculated:", average_rh)
```

```python
# Calculating average relative humidity using a different method
dmethod_rh = valid_rh_values.sum() / len(valid_rh_values)
```

```python
# Displaying calculated average relative humidity using a different method
print("Average Relative Humidity (calculated using a different method):", dmethod_rh)
```

```python
# Comparing manually calculated average relative humidity with the result obtained from average_rh
print("Average Relative Humidity (calculated using a different method):", dmethod_rh)
print("Average Relative Humidity Obtained:", average_rh)
```

```python
# Double checking and comparing the two values
if dmethod_rh == average_rh:
    print("Both values are equal. Calculation successful!")
```

```python
# Importing matplotlib.pyplot for creating a visualization of relative humidity over time
import matplotlib.pyplot as plt
```

```python
# Plotting relative humidity over time
plt.figure(figsize=(10, 6))
plt.plot(valid_rh_values.index, valid_rh_values, color='blue', marker='o', linestyle='-', label='Relative Humidity')
plt.axhline(average_rh, color='red', linestyle='--', label='Average Relative Humidity')  # Add horizontal line for average
plt.title('Relative Humidity Over Time')
plt.xlabel('Time')
plt.ylabel('Relative Humidity (%)')
plt.grid(True)
plt.legend()  # Show legend with labels
plt.show()
```
